## **Visualize the dataset:**

### **Scatter plot**

```python
sns.scatterplot(data=data, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

### **Histograms**

```python
sns.histplot(data=data, x='SepalLengthCm', hue='Species', kde=2)
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
sns.histplot(data=data, x='SepalWidthCm', hue='Species', kde=2)
plt.title('Histogram of Sepal Width')
plt.show()
```

### **Box plots**

```python
sns.boxplot(data=data, x='Species', y='SepalLengthCm')
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

```python
sns.boxplot(data=data, x='Species', y='SepalWidthCm')
plt.title('Box Plot of Sepal Width by Species')
plt.show()
```
